```python
%matplotlib inline
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
data_set = pd.read_csv('kangroo.csv', sep=',')

# n the following data pairs
# X = nasal length (mm ¥10)
# Y = nasal width (mm ¥ 10)
# for a male gray kangaroo from a random sample of such animals.

n_length = pd.DataFrame(data = data_set,columns =['X'])
x = np.array(n_length['X'].values)
n_width = pd.DataFrame(data = data_set,columns =['Y'])
y = np.array(n_width['Y'].values)

slope, intercept, r_value,p_value, std_err = stats.linregress(x,y)

plt.ylabel('nasal width')
plt.xlabel('nasal length')
plt.plot(x,y,'ro',color='black')
plt.plot(x,x*slope+intercept,'g')
plt.axis([400,1000,0,500])
plt.plot()

# Predict width for length
p_length = 600
p_width = p_length*slope+intercept
print p_width

plt.show()


```
